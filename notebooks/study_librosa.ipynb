```python
import librosa, librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Quick starter

filename = librosa.example('nutcracker')
sig, sr = librosa.load(filename) 
print('sr: {:d}'.format(sr))
print(sig[:10], sig.shape)
```

```python
plt.figure(figsize = (15, 10))
librosa.display.waveshow(sig, sr = 22050, alpha = 0.5)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```
